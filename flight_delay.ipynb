### Flight Delay Plotting
    The dataset contains information about average arival delay in minutes each month for 14 airlines in US.  

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# load data 
flight_data = pd.read_csv('./flight_delays.csv')
flight_data.head()
```

```python
#use month column as index
flight_data.set_index('Month', inplace=True)

#change index into string
flight_data.index = flight_data.index.map(str)
flight_data.index
```

```python
#check data
flight_data
```

```python
#handle NaN data
flight_data['US'].replace(np.NaN, 0, inplace=True)
flight_data['US']
```

```python
%matplotlib inline

# using seaborn style
mpl.style.use('seaborn')

font = {'family' : 'lato',
        'weight' : 'bold',
        'size'   : 22}

mpl.rc('font', **font)
```

# Plotting bar chart for AA

```python
# get the data 
data_AA = flight_data['AA']

# change index to int for plotting
data_AA.index = data_AA.index.map(int)

# handle data value < 0 
data_AA[data_AA < 0] = 0
data_AA
```

```python
# show statistics about data_AA
data_AA.describe()
```

```python
# plotting
data_AA.plot(
    kind='bar',
    figsize=(10,6)
)

plt.title('Average Arival Flight Delay for AA in minutes')
plt.ylabel('Minutes')
plt.xticks(rotation=0)

plt.rc


plt.show()
```

# Plotting box chart 

```python
data_AA.plot(
    kind='box',
    figsize=(10,6)
)

plt.title('Box chart for average arrival delay in AA')
plt.ylabel('Minutes')

plt.show()
```
